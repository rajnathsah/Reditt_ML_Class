```python
import statsmodels.api as sm
```

# Load boston data set

```python
from sklearn import datasets
data=datasets.load_boston()
```

# Print description of boston data set

```python
print(data.DESCR)
```

```python
data.feature_names
```

```python
import numpy as np
import pandas as pd
```

# Convert data set into dataframe

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
target=pd.DataFrame(data.target,columns=["MEDV"])
```

# Split data set into independent and dependent variable

```python
X=df["RM"]
y=target["MEDV"]
```

# Fit stats model

```python
model=sm.OLS(y,X).fit()
```

# Predit values

```python
predictions=model.predict(X)
```

# Print model summary

```python
model.summary()
```

# Add constant to independent data set

```python
X=df["RM"]
y=target["MEDV"]
```

```python
X=sm.add_constant(X)
```

# Fit model

```python
model=sm.OLS(y,X).fit()
```

```python
predictions=model.predict(X)
```

# Print model summary

```python
model.summary()
```

# Repeat above with multiple independent variables

```python
X=df[["RM","LSTAT"]]
y=target["MEDV"]
```

```python
model=sm.OLS(y,X).fit()
predictions=model.predict(X)
```
